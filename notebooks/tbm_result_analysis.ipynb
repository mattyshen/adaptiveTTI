```python
#N = 5 # top 3 interactions
def inv_map(ohe_cols, data, N):
    data = ast.literal_eval(data)
    collect = []
    for x in data[:N]:
        collect.append([ohe_cols[int(y)] for y in x])
    return collect

def compute_gt_diff(X_test_t, X_test):
    diff_indices = np.where(X_test_t.to_numpy() != X_test.to_numpy())[0]
    return diff_indices

def compute_edit_effectiveness(gt_diff_indices, concepts, interv_group):
    gt = [concepts[i] for i in gt_diff_indices] # concepts that we should edit
    visited = []
    hit = 0
    total_suggested = set()
    for lst in interv_group:
        new_group = True
        for x in lst:
            concept_name = x.split('_')[0]
            total_suggested.add(concept_name)
            if concept_name in gt and concept_name not in visited:
                if new_group:
                    hit += 1
                    new_group = False
                visited.append(concept_name)
    
    tp = len(visited)
    fp = len(total_suggested) - len(visited)
    fn = len((set(concepts) - total_suggested).union(set(gt)))
    tn = len((set(concepts) - total_suggested - set(gt)))
    
    print(f'{hit}/{len(interv_group)} useful group')
    print(f'acc: {round(len(visited)/len(gt)*100, 2)}')
    print(f'precision: {round(tp/(tp+fp)*100, 2)}')
    print(f'recall: {round(tp/(tp+fn)*100, 2)}')
    
#adap_text = inv_map(ohe_cols, adap[4], 5)
#rand_text = inv_map(ohe_cols, rand[4], 5)
#l_text = inv_map(ohe_cols, l[4], 5)
```

```python
import numpy as np

def parse_concept(x):
    if x[-4] == '-':
        concept_name = x[:-5]
        score = float(x[-4:])
    else:
        concept_name = x[:-4]
        score = float(x[-3:])
    return concept_name

def compute_n_group_edits_to_get_it_right(orig_X_test_t, X_test, interv_methods, true_y):
    res = {}
    for k, lst in interv_methods.items():
        group_cnt = 1
        X_test_t = orig_X_test_t.copy()
        suggested_concepts = set()
        suggested_concepts_in_groups = []
        for group in lst:
            #print(group)
            
            
            suggested_concepts = suggested_concepts.union(set([parse_concept(x) for x in group]))
            suggested_concepts_in_groups.append([parse_concept(x) for x in group])
            
            edit_idxs = [concepts.index(i) for i in suggested_concepts]
            X_test_t.iloc[edit_idxs] = X_test.iloc[edit_idxs]
            counterfactual_pred = model.predict(X_test_t.to_numpy().reshape(1, -1))
            #print(counterfactual_pred)
            if task_name == "agnews":
                if counterfactual_pred[0] == true_y:
                    res[k] = group_cnt
                    print(suggested_concepts_in_groups)
                    break
            else:
                if np.abs(counterfactual_pred[0] - true_y) < 0.5:
                    res[k] = group_cnt
                    print(suggested_concepts_in_groups)
                    break
            group_cnt += 1
    return res

import ast
from sklearn.preprocessing import OneHotEncoder
def remove_dup(input_lst):
        res = []
        for x in input_lst:
            if x not in res:
                res.append(x)
        return res
    
categories = [y for y in concept_df['Response Mapping'].apply(lambda x: remove_dup([str(float(cat)) for cat in ast.literal_eval(x).values()]))]
encoder = OneHotEncoder(categories=categories, handle_unknown='ignore')

encoded_X_train_t = encoder.fit_transform(X_train_teacher.astype(str))
ohe_cols = encoder.get_feature_names_out()
print(len(ohe_cols))
```

```python
# generate data for human to label true concepts
import pandas as pd
from os.path import join

task_name = "agnews" #"agnews", "cebab"
teacher_type = "linear"
saved_data_path = "tbm_gpt4"
label_type = {'cebab': [1, 2, 3, 4, 5], "agnews": [0, 1, 2, 3]}
n = 5 # 5 samples per class

concept_df = pd.read_csv(join(saved_data_path, "{}_{}".format(task_name, teacher_type), "{}_concepts.csv".format(task_name)))
test_df = pd.read_csv(join(saved_data_path, "{}_{}".format(task_name, teacher_type), "{}_test_df.csv".format(task_name)))

df_by_labels = []
label_lst = label_type[task_name]
    
for x in label_lst:
    selected = test_df.loc[test_df['label'] == x].sample(n=n, random_state=42)
    df_by_labels.append(selected)

subset_merged = pd.concat(df_by_labels, ignore_index=True)
subset = subset_merged.sample(frac=1).reset_index(drop=True)

subset.to_csv(join(saved_data_path, "human_annotation", f'{task_name}_{teacher_type}', "full_test_subset.csv"), index=False)
concept_df.to_csv(join(saved_data_path, "human_annotation", f'{task_name}_{teacher_type}', "concepts.csv"), index=False)

concepts = concept_df['Concept Name'].to_list()
cols_to_keep = ['text']+concepts
X_test = subset[cols_to_keep]
X_test.to_csv(join(saved_data_path, "human_annotation", f'{task_name}_{teacher_type}', "X_test.csv"), index=False)
```

```python
# generate data for Matthew to debug why FIGS didn't grow big

def load_teacher_model(task_type, teacher_type, gpu=0):
    ### TODO: load in teacher model using teacher_path ###
    if task_type == "regression":
        if teacher_type == "linear":
            model = LinearRegression()
        elif teacher_type == "mlp1":
            model = MLPRegressor(hidden_layer_sizes=(50,), early_stopping=True, random_state=42)
        elif teacher_type == "mlp2":
            model = MLPRegressor(hidden_layer_sizes=(50,50,), early_stopping=True, random_state=42)
    else:
        if teacher_type == "linear":
            model = LogisticRegression(random_state=42)
        elif teacher_type == "mlp1":
            model = MLPClassifier(hidden_layer_sizes=(50,), early_stopping=True, random_state=42)
        elif teacher_type == "mlp2":
            model = MLPClassifier(hidden_layer_sizes=(50,50,), early_stopping=True, random_state=42)
    
    return model

def generate_tabular_distillation_data(model, saved_data_path, task_name, teacher_type):
    ### TODO: generate teacher train and test data using model, train_path, and test_path ###
    # X_train_teacher is N by c consisting of the "learned" concept vectors of N samples
    # X_train is N by c consisting of "groud-truth" concept vectors of N samples
    # y_test_teacher is N by o consisting of the prediction vectors of N samples
    # y_test is N by 1 consisting of "groud-truth" labels
    
    concept_df = pd.read_csv(join(saved_data_path, "{}_{}".format(task_name, teacher_type), "{}_concepts.csv".format(task_name)))
    train_df = pd.read_csv(join(saved_data_path, "{}_{}".format(task_name, teacher_type), "{}_train_df.csv".format(task_name)))
    test_df = pd.read_csv(join(saved_data_path, "{}_{}".format(task_name, teacher_type), "{}_test_df.csv".format(task_name)))
    
    concepts = concept_df['Concept Name'].to_list()
    
    X_train_teacher = train_df[concepts]
    y_train = train_df['label'].to_numpy().reshape(-1, 1)
    
    X_test_teacher = test_df[concepts]    
    y_test = test_df['label'].to_numpy().reshape(-1, 1)
    
    if 'preds' in train_df:
        y_train_teacher = train_df['preds'].to_numpy().reshape(-1, 1) # some preds were saved
        y_test_teacher = test_df['preds'].to_numpy().reshape(-1, 1)
    else:
        model.fit(X_train_teacher, y_train)
        y_train_teacher = model.predict(X_train_teacher).reshape(-1, 1)
        y_test_teacher = model.predict(X_test_teacher).reshape(-1, 1)
    
    return X_train_teacher, X_test_teacher, y_train_teacher, y_test_teacher, y_train, y_test
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neural_network import MLPRegressor, MLPClassifier
from os.path import join
import os.path

task_type = "regression"
teacher_type = "mlp2"
saved_data_path = "tbm_gpt4"
task_name = "cebab"

teacher = load_teacher_model(task_type, teacher_type)
X_train_t, X_test_t, y_train_t, y_test_t, y_train, y_test = generate_tabular_distillation_data(teacher, saved_data_path, task_name, teacher_type)

obj = {'X_train_t': X_train_t, 'X_test_t': X_test_t, 'y_train_t': y_train_t, 'y_test_t': y_test_t, 'y_train': y_train, 'y_test': y_test}

with open("{}_{}_tbm.pkl".format(task_name,teacher_type), 'wb') as fp:
    pickle.dump(obj, fp)
    
with open("{}_{}_tbm.pkl".format(task_name,teacher_type), 'rb') as fp:
    banana = pickle.load(fp)
print(banana['X_train_t'])
```
