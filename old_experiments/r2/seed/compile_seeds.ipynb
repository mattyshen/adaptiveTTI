```python
import pandas as pd
import numpy as np
import json
```

```python
for t in ['true', 'hat']:    
    for model in ['bin_depth3', 'bit_depth3', 'bit_depth4']:
        dfs = []
        for i in range(5):
            dfs.append(pd.read_csv(f'{model}_distillation_{t}_R2_seed{i}.csv', index_col = 0).set_index('Model'))

        concatenated_df = np.stack([df.values for df in dfs])

        # Calculate mean and standard deviation along the new axis
        mean_df = np.mean(concatenated_df, axis=0)
        std_df = np.std(concatenated_df, axis=0)

        # Create a new dataframe to store the formatted results
        formatted_df = pd.DataFrame(index=dfs[0].index, columns=dfs[0].columns)

        # Fill in the new dataframe with mean (std) format
        for i, col in enumerate(dfs[0].columns):
            for j, row in enumerate(dfs[0].index):
                mean_value = mean_df[j, i]
                std_value = std_df[j, i]
                formatted_df.at[row, col] = f"{mean_value:.3f} ({std_value:.3f})"

        # Display the formatted dataframe
        formatted_df.to_csv(f'complete/{model}_{t}_R2.csv')
```
